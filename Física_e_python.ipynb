##Relacionais

````python
x = 10
y = 7
z = 3
# >maior que "greater than" 
print(x>y) #retorna um tipo booleano#
print(x>y>z)

# < menor que "less than"
print(x<y)
print(y<x)

# == Igual a "equal to"
print(x==y)
print(x==10) #pode comparar a um valor numérico

# != não igual (diferente)
print(x != y)
print(x != 10)

# >= maior ou igual a "greater than or equal to"
print(x >= y)
print(x >= 10)

# <= meno ou igual a "less than or equal to "
print(x <= y)

````

````python
#Atribuição direta

a = 7
b = 2

print(a,b)

a += b
print(a,b)

a = 7
b = 2

a -= b #a = a - b 
print(a,b)

a = 7
b = 2

a *= b # a = a * b
print(a,b)

a = 7 
b = 2

a /= b # a = a/b
print(a,b)
````
